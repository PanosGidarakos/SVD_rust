```python
import proto 
import cheapgpt
import numpy.linalg as ling
import wholeexp
import last 
import numpy as np
import pandas as pd

np.random.seed(0)
```

    Frobenius norm

```python
for ms in [10,20,30,40]: 
    A_example = np.random.rand(ms, ms)
    print(f'Array{ms}x{ms}')
    print()
    print('first sigmna,U,V')
    Sigma, U, V = cheapgpt.golub_reinsch_svd(A_example)
    norm = np.linalg.norm(A_example - np.dot(U, np.dot(Sigma, np.transpose(V))), 'fro')
    print(norm)
    print()
    print('Cheap gtp sigmna,U,V')
    Sigma, U, V = last.svd(A_example)
    norm = np.linalg.norm(A_example - np.dot(U, np.dot(Sigma, np.transpose(V))), 'fro')
    print(norm)


    print()
    print('numpy svd U,S Vh')
    U,Sigma,V=ling.svd(A_example,full_matrices=False)
    norm = np.linalg.norm(A_example - np.dot(U * Sigma, V), 'fro')
    print(norm)
    print()

    print('last gpt U S Vt')
    U, Sigma, V = wholeexp.svd(A_example)
    norm = np.linalg.norm(A_example - np.dot(U * np.diag(Sigma), V), 'fro')
    print(norm)
    print()
```
